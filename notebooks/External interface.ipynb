```python
import pandas as pd
import json

# 函数名改为更准确的 extract_interfaces_to_json
def extract_interfaces_to_json(excel_path, output_json_path):
    # 读取 Excel 文件
    # header=1 表示将 Excel 的第 2 行（索引为1）作为列名（即包含 Cellular, WiFi 等的那一行）
    df = pd.read_excel(excel_path, header=1)
    
    # 清理列名中的换行符和多余的空格（Excel中经常会有各种隐藏字符）
    df.columns = [str(col).replace('\n', ' ').strip() for col in df.columns]
    
    # 获取所有的接口列名（从第3列开始到最后一列）
    interface_columns = df.columns[2:] 
    
    results = [] # results 本身就是一个 Python 列表
    
    # 遍历每一行数据
    for index, row in df.iterrows():
        component_name = str(row['零部件名称']).strip()
        
        # 如果零部件名称为空或者是 NaN，则跳过（比如表格底部的空白行）
        if pd.isna(row['零部件名称']) or not component_name:
            continue
            
        active_interfaces = []
        
        # 检查每个接口列
        for col in interface_columns:
            cell_value = str(row[col]).strip()
            # 如果单元格内容包含 '√'，则将该列名加入接口列表
            if '√' in cell_value:
                active_interfaces.append(col)
        
        # 只有当存在外部接口时，才加入结果列表
        if active_interfaces:
            results.append({
                "component": component_name,
                "interfaces": active_interfaces
            })

    # ================= 核心修改点 =================
    # 直接将整个 results 列表一次性写入为标准的 JSON 数组格式
    # ensure_ascii=False 保证中文正常显示，indent=4 让输出的 JSON 有良好的缩进，方便阅读
    with open(output_json_path, 'w', encoding='utf-8') as f:
        json.dump(results, f, ensure_ascii=False)
    # =============================================

    print(f"提取完成！共提取了 {len(results)} 个零部件的接口信息，已保存至 {output_json_path}")

# 使用示例
# 输出文件的后缀记得从 .jsonl 改为 .json
extract_interfaces_to_json(
    r'C:\Users\18056\OneDrive\Desktop\研一\nlp\汽车模型\TARA\数据\E16-EU项目对外接口清单.xlsx', 
    r'C:\Users\18056\OneDrive\Desktop\研一\nlp\汽车模型\TARA\数据\information\output.json'
)
```
